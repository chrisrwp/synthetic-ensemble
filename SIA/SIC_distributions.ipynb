```python
month_ = 9
lat_min = 75
lat_max = 85

SIC_distributions = {}
for model_i, model_name in enumerate(model_names):
    print(model_name)
    if model_name == 'MPI_ESM1':
        lat_name = 'latitude'
    else:
        lat_name = 'lat'
   
    reduced_SIC = xr.open_dataarray('/glade/campaign/univ/ucub0084/Synthetic_ensemble/SIC/Reduced_datasets/'\
                                      +'{}_reduced_1979-2020_0{}.nc'.format(model_name,month_))

    areacello = xr.open_dataset('/glade/campaign/univ/ucub0084/Synthetic_ensemble/Raw_data/areacello/areacello_{}_30N.nc'.format(model_name))
    
    if model_name in ['CanESM2','CSIRO_MK36']:
        SIC_selected = reduced_SIC.where(reduced_SIC['lat']>lat_min)
    else:
        SIC_selected = reduced_SIC.load() * (areacello['areacello'].where(areacello[lat_name]>lat_min,drop=True)*0 + 1)
    
    SIC_distributions[model_name] = SIC_selected
```

```python
reduced_SIC.where(xr.ufuncs.logical_not(xr.ufuncs.isnan(areacello['areacello'].where(areacello[lat_name]>lat_min,drop=True)))).sel(member=1).isel(time=0).plot()
```
